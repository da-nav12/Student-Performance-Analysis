### Correlation matrix

```python

plt.figure(figsize=(9, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

### Distribution plots for numeric columns

```python

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[numeric_cols].hist(figsize=(12, 10), bins=20)
plt.tight_layout()
plt.suptitle("Distribution of Numerical Features", y=1.02)
plt.show()
```

### Get categorical columns

```python

import math 
cat_cols = df.select_dtypes(include='object').columns
n = len(cat_cols)

# Determine grid size (square or nearly square)
cols = 3
rows = math.ceil(n / cols)

# Set up subplots
fig, axes = plt.subplots(rows, cols, figsize=(10, 5 * rows))
axes = axes.flatten()  # Flatten in case of 1D

# Generate pie charts
for i, col in enumerate(cat_cols):
    df[col].value_counts().plot.pie(
        ax=axes[i],
        autopct='%1.1f%%',
        startangle=90,
        colors=sns.color_palette('Set3')
    )
    axes[i].set_title(f"Proportion of {col}")
    axes[i].set_ylabel('')

# Hide any extra subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# Select numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Plot each numeric column as a line
plt.figure(figsize=(12, 6))

for col in numeric_cols:
    plt.plot(df[col].values, label=col, linewidth=2.5)

plt.title("Pair Line Plot of Numeric Features")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
